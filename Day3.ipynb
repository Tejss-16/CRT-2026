#recurrsion

```python
def exp(i,n):
  if i== n+1:
    return
  print(i)
  exp(i+1,n)
exp(1,5)

```

```python
def sum(n,m=0):
  if n==0:
    print(m)
    return
  sum(n-1,m+n)
sum(2)
```

```python
def fac(n,m=1):
  if n==0:
    print(m)
    return
  fac(n-1,m*n)
fac(2)
```

```python
def fac(n):
  if n==0:
    return 1
  return n * fac(n-1)
fac(3)
```

```python
def per(n,r):
  fact=1
  for i in range (1,n+1):
    fact*=i
  fact2=1
  l=n-r
  for j in range (1,l+1):
    fact2*=j
  print(fact/fact2)
per(7,4)
```

```python
class student:
  name ="tejas"
  def __init__(self, sub1, sub2, sub3):
    self.m1 = sub1
    self.m2 = sub2
    self.m3 = sub3
    self.avg = (self.m1 + self.m2 + self.m3)/3

s1 = student(80, 85, 90)
print(f"The average for s1 is: {s1.avg}")
```

```python
class student:
    sclName="vcet"
    def __init__(self,name,marks):
        self.Name = name
        self.M =marks

    def display_details(self):
        print(student.sclName,self.Name,self.M)
    def  is_pass(self):
        return self.M >=10


s1 = student("Tejas",20)
s1.display_details()
print()
print("pass: ",s1.is_pass())
```

```python
class employee:
  def __init__(self,role,department, salary):
    self.R =role
    self.D =department
    self.S =salary

  def show_details(self):
    print(self.R)
    print(self.D)
    print(self.S)
s1 = employee("CEO","XYZ",200000)
s1.show_details()

print()
class engineer(employee):
  def __init__(self,name,age,role,department,salary):
    self.N =name
    self.A =age
    super().__init__(role,department,salary)

  def display(self):
    print(self.N)
    print(self.A)
    super().show_details()

s1 = engineer("tejas",20,"intern","HR",200000)
s1.display()
```
