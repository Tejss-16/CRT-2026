# While loop

```python
num = 12345
count=0
while num>0:
  count+=1
  num=num//10
print(count)
```

```python
# print sum of number
num=1234
sum=0
while num!=0:
  rem=num%10
  sum+=rem
  num=num//10
print(sum)
```

```python
num = 1234
rev=[]
while num!=0:
  rem = num%10
  rev.append(rem)
  num=num//10
print(*rev)
```

```python
num = 121
n=num
rev=0
while num!=0:
  rem = num%10
  rev=rev*10+rem
  num=num//10
if rev==n:
  print("palindrom")
else:
  print("not")
```
